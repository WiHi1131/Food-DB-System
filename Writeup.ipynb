## Database Creation

The 'db_create' route of my Flask application (shown in the code snippet below) plays a pivotal role in setting up the database for my project. This route encapsulates the essence of my understanding and application of SQL in database creation and management. Utilizing the psycopg2 library to interface with PostgreSQL, this route executes a series of SQL commands to construct the fundamental structure of my database.

- Table Creation:

    - Foods Table: The first SQL command in this route creates the Foods table, which serves as a repository for various food items. The table includes fields such as food_id, name, and various nutritional attributes. The SERIAL PRIMARY KEY and NOT NULL UNIQUE constraints ensure data integrity by providing unique identifiers for each food item and preventing null entries for the food name.
    - Dishes Table: Next, the Dishes table is created to store information about different dishes. Similar to the Foods table, it has a SERIAL PRIMARY KEY and includes a unique name and a description.
    - FoodsInDish Table: This table represents a junction between Foods and Dishes, linking food items to the dishes they compose. It includes FOREIGN KEY constraints referencing Foods and Dishes, with ON DELETE CASCADE to maintain referential integrity.
    - Meals, Users, and Days Tables: These tables further extend the database schema, incorporating aspects like meal times, user information, and daily logs. Each table includes foreign keys with ON DELETE CASCADE to automatically handle the deletion of dependent records.
    - FoodUpdatesLog Table: This table is designed to log changes in the Foods table, capturing both old and new values in JSONB format, facilitating comprehensive tracking of updates.
    
- Index Creation:

    - An index is created on the name column of the Foods table (idx_foods_name), optimizing searches and queries based on the food name.
    
- Trigger Functions and Triggers:

    - The db_create route also includes the creation of trigger functions and triggers. These are crucial for enforcing specific business rules and logging mechanisms.
    - Prevent Name Update Triggers: For both the Foods and Dishes tables, triggers are set up to prevent updates to the name column, ensuring that the names of foods and dishes remain constant once created.
    - Log Food Updates Trigger: This trigger captures changes in nutritional information in the Foods table, logging them into the FoodUpdatesLog table. It showcases a dynamic response to data modifications, a critical aspect of database management.

````python
@app.route('/db_create')
def create():
    conn = psycopg2.connect("postgres://food_db_msqq_user:96WkFN4LYyA6g0p8n9ykbw7GT0KQudsM@dpg-clok7g1oh6hc73bia110-a/food_db_msqq")
    cur = conn.cursor()
    # Create Foods table if it doesn't exist
    cur.execute('''
        CREATE TABLE IF NOT EXISTS Foods (
        food_id SERIAL PRIMARY KEY,
        name TEXT NOT NULL UNIQUE,
        
        portion_size REAL,           -- Usually in grams (g)
        calories REAL,               -- Usually in kcal
        total_fat REAL,              -- Typically in grams (g)
        saturated_fat REAL,          -- Typically in grams (g)
        trans_fat REAL,              -- Typically in grams (g)
        cholesterol REAL,            -- Typically in milligrams (mg)
        sodium REAL,                 -- Typically in milligrams (mg)
        total_carbohydrates REAL,    -- Typically in grams (g)
        dietary_fiber REAL,          -- Typically in grams (g)
        sugars REAL,                 -- Typically in grams (g)
        protein REAL,                -- Typically in grams (g)
        vitamin_d REAL,              -- Usually in micrograms (Âµg) or IU
        calcium REAL,                -- Typically in milligrams (mg)
        iron REAL,                   -- Typically in milligrams (mg)
        potassium REAL               -- Typically in milligrams (mg)
        );
    ''')
    #Create index on name of Foods table
    cur.execute('''
        CREATE INDEX idx_foods_name ON Foods (name);
        ''')
    # Create Dishes table
    cur.execute('''
        CREATE TABLE IF NOT EXISTS Dishes (
        dish_id SERIAL PRIMARY KEY,
        name TEXT NOT NULL UNIQUE,
        description TEXT
        );
    ''')
    
    # Create foodsinDish table (junction table)
    cur.execute('''
        CREATE TABLE IF NOT EXISTS foodsInDish (
        food_in_dish_id SERIAL PRIMARY KEY,
        dish_id INT,
        food_id INT,
        portion_multiplier REAL,  -- Multiplier for the portion size from Foods table
        FOREIGN KEY (dish_id) REFERENCES Dishes(dish_id) ON DELETE CASCADE,
        FOREIGN KEY (food_id) REFERENCES Foods(food_id) ON DELETE CASCADE
        );
    ''')
    
    # Create an Enum type for meal_type
    cur.execute('''
        DO $$ BEGIN
            CREATE TYPE meal_type AS ENUM ('breakfast', 'lunch', 'dinner', 'snack');
        EXCEPTION
            WHEN duplicate_object THEN null;
        END $$;
    ''')
    
    cur.execute('''
        CREATE TABLE IF NOT EXISTS Meals (
        meal_id SERIAL PRIMARY KEY,
        meal_time TIMESTAMP NOT NULL,
        dish_id INT,                 -- Foreign key to reference the Dishes table
        type meal_type NOT NULL,       -- Enum type for meal type
        FOREIGN KEY (dish_id) REFERENCES Dishes(dish_id) ON DELETE CASCADE
        );
    ''')
    
    # Create Users table
    cur.execute('''
        CREATE TABLE IF NOT EXISTS Users (
        user_id SERIAL PRIMARY KEY,
        username TEXT NOT NULL UNIQUE
        );
    ''')

    # Create Days table
    cur.execute('''
        CREATE TABLE IF NOT EXISTS Days (
        day_id SERIAL PRIMARY KEY,
        date DATE NOT NULL,
        user_id INT,
        meal_id INT,
        FOREIGN KEY (user_id) REFERENCES Users(user_id) ON DELETE CASCADE, 
        FOREIGN KEY (meal_id) REFERENCES Meals(meal_id) ON DELETE CASCADE
        );
    ''')
    
    # Create FoodUpdatesLog
    cur.execute('''
        CREATE TABLE IF NOT EXISTS FoodUpdatesLog (
        log_id SERIAL PRIMARY KEY,
        food_id INT,
        old_values JSONB,
        new_values JSONB,
        update_time TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        FOREIGN KEY (food_id) REFERENCES Foods(food_id) ON DELETE CASCADE
    );
    ''')
    
    def create_triggers(cur):
        # Trigger function to prevent updates to the name in Foods
        cur.execute('''
            CREATE OR REPLACE FUNCTION prevent_name_update_foods()
            RETURNS TRIGGER AS $$
            BEGIN
                IF OLD.name IS DISTINCT FROM NEW.name THEN
                    RAISE EXCEPTION 'Updates to food name are not allowed.';
                END IF;
                RETURN NEW;
            END;
            $$ LANGUAGE plpgsql;
        ''')
        
        cur.execute('''
            CREATE TRIGGER prevent_name_update_foods_trigger
            BEFORE UPDATE OF name ON Foods
            FOR EACH ROW
            EXECUTE FUNCTION prevent_name_update_foods();
        ''')
        
        # Trigger function to prevent updates to the name in Dishes
        cur.execute('''
            CREATE OR REPLACE FUNCTION prevent_name_update_dishes()
            RETURNS TRIGGER AS $$
            BEGIN
                IF OLD.name IS DISTINCT FROM NEW.name THEN
                    RAISE EXCEPTION 'Updates to dish name are not allowed.';
                END IF;
                RETURN NEW;
            END;
            $$ LANGUAGE plpgsql;
        ''')
        
        cur.execute('''
            CREATE TRIGGER prevent_name_update_dishes_trigger
            BEFORE UPDATE OF name ON Dishes
            FOR EACH ROW
            EXECUTE FUNCTION prevent_name_update_dishes();
        ''')
        #Trigger function to log updates in Foods
        cur.execute('''
            CREATE OR REPLACE FUNCTION log_food_updates()
            RETURNS TRIGGER AS $$
            BEGIN
                INSERT INTO FoodUpdatesLog (food_id, old_values, new_values)
                VALUES (
                    OLD.food_id,
                    jsonb_build_object(
                        'portion_size', OLD.portion_size,
                        'calories', OLD.calories,
                        'total_fat', OLD.total_fat,
                        'saturated_fat', OLD.saturated_fat,
                        'trans_fat', OLD.trans_fat, 
                        'cholesterol', OLD.cholesterol,
                        'sodium', OLD.sodium, 
                        'total_carbohydrates', OLD.total_carbohydrates,
                        'dietary_fiber', OLD.dietary_fiber, 
                        'sugars', OLD.sugars,
                        'protein', OLD.protein, 
                        'vitamin_d', OLD.vitamin_d, 
                        'calcium', OLD.calcium,
                        'iron', OLD.iron,
                        'potassium', OLD.potassium
                    ),
                    jsonb_build_object(
                        'portion_size', NEW.portion_size,
                        'calories', NEW.calories,
                        'total_fat', NEW.total_fat,
                        'saturated_fat', NEW.saturated_fat,
                        'trans_fat', NEW.trans_fat, 
                        'cholesterol', NEW.cholesterol,
                        'sodium', NEW.sodium, 
                        'total_carbohydrates', NEW.total_carbohydrates,
                        'dietary_fiber', NEW.dietary_fiber, 
                        'sugars', NEW.sugars,
                        'protein', NEW.protein, 
                        'vitamin_d', NEW.vitamin_d, 
                        'calcium', NEW.calcium,
                        'iron', NEW.iron,
                        'potassium', NEW.potassium
                    )
                );
                RETURN NEW;
            END;
            $$ LANGUAGE plpgsql;
        ''')
        #Trigger for logging updates
        cur.execute('''
            CREATE TRIGGER foods_update_log_trigger
            AFTER UPDATE OF calories, total_fat, sodium -- Add other fields as needed
            ON Foods
            FOR EACH ROW
            WHEN (OLD.* IS DISTINCT FROM NEW.*)
            EXECUTE FUNCTION log_food_updates();
        ''')
        
    
    # Create triggers
    create_triggers(cur)
    
    conn.commit()
    conn.close()
    return "All Tables Successfully Created!"
````

## Selecting from Database

The 'db_select' route in my Flask application (in code shown in the snippet below) is instrumental in showcasing the retrieval and presentation of data stored in the database. This route utilizes SQL `SELECT` statements to fetch data from each table and display it in a web format. Here's an overview of the SQL used:

1. **Retrieval of Data from Each Table**:
   - The core functionality of this route is to execute `SELECT * FROM {table_name};` for each table in the database. This SQL statement retrieves all records from the specified table.
   - The tables included are `Foods`, `Dishes`, `foodsinDish`, `Meals`, `Users`, and `Days`, covering the entire spectrum of data stored in the database.

2. **Fetching and Formatting Records**:
   - For each table, the `cur.execute` method is called with the SQL query to fetch all records.
   - The `fetchall` method is then used to retrieve all rows of the query result, which are stored in the variable `records`.
   - Additionally, `cur.description` is used to extract column names for each table, which are important for presenting the data in an understandable format.

3. **HTML Table Generation**:
   - The retrieved data, along with the column names, are passed to the `format_records_as_table` function, which dynamically generates an HTML table for each set of records.
   - This function creates a string of HTML, starting with a header for the table name, followed by the creation of table headers (`<th>`) for each column, and then iterates over each record to populate the table rows (`<tr>`) with data (`<td>`).

4. **Concatenating Results for Web Display**:
   - The HTML strings for each table are concatenated into a single response string, which is then returned to be displayed on the webpage.

````python
@app.route('/db_select')
def selecting():
    conn = psycopg2.connect("postgres://food_db_msqq_user:96WkFN4LYyA6g0p8n9ykbw7GT0KQudsM@dpg-clok7g1oh6hc73bia110-a/food_db_msqq")
    cur = conn.cursor()
    
    # Function to format records into an HTML table
    def format_records_as_table(records, table_name, column_names):
        response_string = "<h2>{}</h2>".format(table_name)
        response_string += "<table border='1'>"
        
        # Add column headers
        response_string += "<tr>"
        for col_name in column_names:
            response_string += "<th>{}</th>".format(col_name)
        response_string += "</tr>"
        
        # Add table data
        for record in records:
            response_string += "<tr>"
            for info in record:
                response_string += "<td>{}</td>".format(info)
            response_string += "</tr>"
        response_string += "</table><br>"
        return response_string

    def query_and_format_table(cur, table_name):
        cur.execute(f'SELECT * FROM {table_name};')
        records = cur.fetchall()
        column_names = [desc[0] for desc in cur.description]
        return format_records_as_table(records, table_name, column_names)

    # Query and display data from each table
    response_string = ""
    for table in ["Foods", "Dishes", "foodsinDish", "Meals", "Users", "Days"]:
        response_string += query_and_format_table(cur, table)

    conn.close()
    return response_string
````

## Frontend Food Lookup

In my Flask application, two routes specifically utilize SQL for querying the database to provide food-related information, which are crucial for the working frontend functionality, allowing users to input and find foods that are within the database. Here's a summary of the SQL statements I used:

1. **SQL in the `/get_food_suggestions` Route**:
   - This route was designed to provide autocomplete suggestions for food names as users type in a search term.
   - The SQL query that was used is: `"SELECT name FROM Foods WHERE name ILIKE %s LIMIT 10"`.
   - This statement is crafted to select the `name` field from the `Foods` table, where the name matches the user-provided search term, using the `ILIKE` operator for case-insensitive matching.
   - The search term is appended with `%` on both sides to enable a pattern match where the search term can appear anywhere within the food name.
   - The `LIMIT 10` clause in this query limits the results to a maximum of 10 entries, ensuring the response is manageable and suitable for autocomplete suggestions.

2. **SQL in the `/get_food_info` Route**:
   - In this route, detailed nutritional information is displayed about a specific food item when a user selects a food name.
   - The SQL query employed is: `"SELECT * FROM Foods WHERE name = %s"`.
   - This statement selects all fields (`*`) from the `Foods` table for the food item whose name exactly matches (`=`) the provided food name.
   - As this query is expected to return only one record (since food names are unique),  `fetchone()` is used to retrieve the single matching record.

## Basic Queries 

In the `/db_queries` route of the project, a series of SQL queries are executed to demonstrate basic data retrieval and aggregation capabilities. This route is key for showcasing the ability to perform various types of queries on the database. Here's an overview of the SQL statements used (relevant code in the snippet below) :

- Query for Foods with Nonzero Cholesterol
    - This query retrieves all records from the Foods table where the cholesterol value is greater than zero. It's an example of a basic conditional selection.
- Query for Meals Eaten Before 13:00
    - This query selects all records from the Meals table where the meal time is before 13:00 (1 PM). It utilizes the EXTRACT function to retrieve the hour part from the meal_time timestamp.
- Query for Days Logged by User ID 1
    - This statement fetches all entries in the Days table corresponding to user_id 1. It's a straightforward query filtering records based on a specific user ID.
- Query to Count Dishes for Each Type of Meal
    - This query demonstrates the use of the GROUP BY clause and the aggregation function COUNT. It groups the records in the Meals table by meal type (breakfast, lunch, etc.) and counts the total number of dishes for each meal type.

The results of these queries are formatted into HTML tables and displayed on the webpage, providing a clear and user-friendly presentation of the data.

````python
@app.route('/db_queries')
def db_queries():
    # Database connection
    conn = psycopg2.connect("postgres://food_db_msqq_user:96WkFN4LYyA6g0p8n9ykbw7GT0KQudsM@dpg-clok7g1oh6hc73bia110-a/food_db_msqq")
    cur = conn.cursor()

    # Function to format records into an HTML table
    def format_records_as_table(records, table_name, column_names):
        response_string = f"<h2>{table_name}</h2>"
        response_string += "<table border='1'>"
        
        # Add column headers
        response_string += "<tr>"
        for col_name in column_names:
            response_string += f"<th>{col_name}</th>"
        response_string += "</tr>"
        
        # Add table data
        for record in records:
            response_string += "<tr>"
            for info in record:
                response_string += f"<td>{info}</td>"
            response_string += "</tr>"
        response_string += "</table><br>"
        return response_string

    # Query for foods with nonzero cholesterol
    cur.execute("SELECT * FROM Foods WHERE cholesterol > 0;")
    foods_records = cur.fetchall()
    foods_columns = [desc[0] for desc in cur.description]
    response_string = format_records_as_table(foods_records, "Foods with Nonzero Cholesterol", foods_columns)

    # Query for meals eaten before 13:00
    cur.execute("SELECT * FROM Meals WHERE EXTRACT(HOUR FROM meal_time) < 13;")
    meals_records = cur.fetchall()
    meals_columns = [desc[0] for desc in cur.description]
    response_string += format_records_as_table(meals_records, "Meals Eaten Before 13:00", meals_columns)

    # Query for days logged by user_id 1
    cur.execute("SELECT * FROM Days WHERE user_id = 1;")
    days_records = cur.fetchall()
    days_columns = [desc[0] for desc in cur.description]
    response_string += format_records_as_table(days_records, "Days Logged by User ID 1", days_columns)
    
    # Query: Count dishes for each type of meal
    cur.execute('''
        SELECT type, COUNT(*) AS total_dishes
        FROM Meals
        GROUP BY type;
    ''')
    meal_type_count = cur.fetchall()
    meal_type_count_columns = [desc[0] for desc in cur.description]
    response_string += format_records_as_table(meal_type_count, "Total Dishes for Each Meal Type (GROUP BY EXAMPLE)", meal_type_count_columns)

    # Close the database connection
    cur.close()
    conn.close()

    return response_string
````

## Joins

The `db_joins` route in the project showcases the use of SQL JOIN operations to perform complex queries involving multiple tables. This route is crucial for demonstrating how different tables in the database can be interconnected to extract meaningful insights. Here's a summary of the SQL statements used in this route (code in snippet below):

- Query for Dishes Containing Spinach: 
    - This query retrieves the names and descriptions of dishes that contain spinach. It involves a JOIN between Dishes, foodsInDish, and Foods tables and filters for dishes where one of the ingredients is spinach.
- Query to Calculate Total Calories in Each Dish:
    - This query computes the total calories for each dish. It multiplies the calories of each food item by its portion multiplier in the dish and then sums these values. The JOIN operation links Dishes, foodsInDish, and Foods tables, and the results are grouped by dish names.
- Query to Find Foods in Lunch Dishes:
    - This query lists the food items used in dishes classified as lunch. It involves multiple JOIN operations across Meals, Dishes, foodsInDish, and Foods tables, filtered for meals of the 'lunch' type.
- Query to Display Dishes by Single Ingredients:
    - This query provides a list of dishes ordered by ingredients. It joins the Foods, foodsInDish, and Dishes tables and orders the results by food name and dish name, offering a clear view of which dishes contain which ingredients.
    
The results from these queries are formatted into HTML tables and displayed on the webpage, making the data easily accessible and understandable to users. 

````python
@app.route('/db_joins')
def db_joins():
    conn = psycopg2.connect("postgres://food_db_msqq_user:96WkFN4LYyA6g0p8n9ykbw7GT0KQudsM@dpg-clok7g1oh6hc73bia110-a/food_db_msqq")
    cur = conn.cursor()

    # Function to format records into an HTML table
    def format_records_as_table(records, table_name, column_names):
        response_string = f"<h2>{table_name}</h2>"
        response_string += "<table border='1'>"
        
        # Add column headers
        response_string += "<tr>"
        for col_name in column_names:
            response_string += f"<th>{col_name}</th>"
        response_string += "</tr>"
        
        # Add table data
        for record in records:
            response_string += "<tr>"
            for info in record:
                response_string += f"<td>{info}</td>"
            response_string += "</tr>"
        response_string += "</table><br>"
        return response_string

    # Query 1: Dishes containing spinach
    cur.execute('''
        SELECT DISTINCT Dishes.name, Dishes.description
        FROM Dishes
        JOIN foodsInDish ON Dishes.dish_id = foodsInDish.dish_id
        JOIN Foods ON foodsInDish.food_id = Foods.food_id
        WHERE Foods.name = 'Spinach';
    ''')
    dishes_spinach = cur.fetchall()
    dishes_spinach_columns = [desc[0] for desc in cur.description]
    response_string = format_records_as_table(dishes_spinach, "Dishes Containing Spinach", dishes_spinach_columns)

    # Query 2: Calories in each dish
    cur.execute('''
        SELECT Dishes.name, SUM(Foods.calories * foodsInDish.portion_multiplier) AS total_calories
        FROM Dishes
        JOIN foodsInDish ON Dishes.dish_id = foodsInDish.dish_id
        JOIN Foods ON foodsInDish.food_id = Foods.food_id
        GROUP BY Dishes.name;
    ''')
    dish_calories = cur.fetchall()
    dish_calories_columns = [desc[0] for desc in cur.description]
    response_string += format_records_as_table(dish_calories, "Calories in Each Dish (GROUP BY EXAMPLE)", dish_calories_columns)

    # Query 3: Foods in lunch dishes
    cur.execute('''
        SELECT Foods.name
        FROM Meals
        JOIN Dishes ON Meals.dish_id = Dishes.dish_id
        JOIN foodsInDish ON Dishes.dish_id = foodsInDish.dish_id
        JOIN Foods ON foodsInDish.food_id = Foods.food_id
        WHERE Meals.type = 'lunch';
    ''')
    lunch_foods = cur.fetchall()
    lunch_foods_columns = [desc[0] for desc in cur.description]
    response_string += format_records_as_table(lunch_foods, "Foods in Lunch Dishes", lunch_foods_columns)
    
    # New Query: Display dishes by single ingredients
    cur.execute('''
        SELECT Foods.name AS Ingredient, Dishes.name AS Dish
        FROM Foods
        JOIN foodsInDish ON Foods.food_id = foodsInDish.food_id
        JOIN Dishes ON foodsInDish.dish_id = Dishes.dish_id
        ORDER BY Foods.name, Dishes.name;
    ''')
    dishes_by_ingredients = cur.fetchall()
    dishes_by_ingredients_columns = [desc[0] for desc in cur.description]
    response_string += format_records_as_table(dishes_by_ingredients, "Dishes Ordered By Ingredients", dishes_by_ingredients_columns)

    cur.close()
    conn.close()
    return response_string
````

## Showing Trigger Execution on Updates

The `/db_updates` route in the project is designed to demonstrate the functionality of triggers set up to regulate updates in the database, specifically within the `Foods` and `Dishes` tables, and to showcase the logging of changes in nutritional values. Here's an analysis of the SQL statements used in this route (code in snippet below, also reference the code used to create the database, where the triggers are also created):

- Attempt to Update Name in Foods Table:
    - This statement tries to update the name of a food item in the Foods table. However, due to a trigger previously set up (prevent_name_update_foods_trigger), this update should fail. The trigger is designed to prevent changes to the name column in the Foods table, ensuring that food names remain constant.
- Attempt to Update Name in Dishes Table:
    - Similarly, this statement attempts to update the name of a dish in the Dishes table. The trigger prevent_name_update_dishes_trigger is expected to block this update, maintaining the integrity of dish names.
- Update Nutritional Values in Foods Table:
    - This statement successfully updates a nutritional value (calories) in the Foods table. Unlike the name, nutritional values are allowed to be updated.
- Display FoodUpdatesLog:
    - After the nutritional update, this query retrieves all records from the FoodUpdatesLog table. This table is populated by another trigger (foods_update_log_trigger), which logs changes in the nutritional values of the Foods table. The log includes the old and new values, providing a transparent record of updates.
    
Each SQL operation is wrapped in a try-except block to handle exceptions, especially for the attempted name updates, which are expected to fail due to the triggers. The successful update of nutritional values and the subsequent retrieval of logs from the FoodUpdatesLog table demonstrate how triggers can be used to enforce business rules (like immutable names) and to automatically log significant changes in data.

````python
@app.route('/db_updates')
def db_updates():
    conn = psycopg2.connect("postgres://food_db_msqq_user:96WkFN4LYyA6g0p8n9ykbw7GT0KQudsM@dpg-clok7g1oh6hc73bia110-a/food_db_msqq")
    cur = conn.cursor()

    def format_records_as_table(records, table_name, column_names):
        response_string = f"<h2>{table_name}</h2>"
        response_string += "<table border='1'>"
        for col_name in column_names:
            response_string += f"<th>{col_name}</th>"
        response_string += "</tr>"
        for record in records:
            response_string += "<tr>"
            for info in record:
                response_string += f"<td>{info}</td>"
            response_string += "</tr>"
        response_string += "</table><br>"
        return response_string

    response_string = ""

    # Attempt to update name in Foods table (should fail)
    try:
        cur.execute("UPDATE Foods SET name = 'Updated Name' WHERE food_id = 1;")
        conn.commit()
        response_string += "<p>Update to Foods name successful (unexpected).</p>"
    except psycopg2.Error as e:
        response_string += f"<p>Error updating Foods name: {e}</p>"
        conn.rollback()

    # Attempt to update name in Dishes table (should fail)
    try:
        cur.execute("UPDATE Dishes SET name = 'Updated Name' WHERE dish_id = 1;")
        conn.commit()
        response_string += "<p>Update to Dishes name successful (unexpected).</p>"
    except psycopg2.Error as e:
        response_string += f"<p>Error updating Dishes name: {e}</p>"
        conn.rollback()

    # Update nutritional values in Foods table and check log
    try:
        cur.execute("UPDATE Foods SET calories = 500 WHERE food_id = 1;")
        conn.commit()
        response_string += "<p>Nutritional values in Foods updated successfully.</p>"

        # Display FoodUpdatesLog
        cur.execute("SELECT * FROM FoodUpdatesLog;")
        log_records = cur.fetchall()
        log_columns = [desc[0] for desc in cur.description]
        response_string += format_records_as_table(log_records, "FoodUpdatesLog", log_columns)
    except psycopg2.Error as e:
        response_string += f"<p>Error updating Foods nutritional values: {e}</p>"
        conn.rollback()

    cur.close()
    conn.close()
    return response_string
````

## Deletes and Cascades

The `/db_deletes` route in the project is designed to demonstrate the impact of deletion operations, especially in the context of foreign key constraints and cascade triggers. This route not only shows what data has been deleted but also the effect of these deletions on related tables. Here's a summary of the SQL statements used (full code snippet below, and DELETE ON CASCADE statements in table creation discussed above):

1. **Deletion Operations Across Tables**:
   - The route follows a specific order of deletions to respect foreign key constraints: `['Days', 'Meals', 'foodsinDish', 'Dishes', 'Foods', 'Users']`.
   - For each table, the SQL statement executed is:
     ```sql
     DELETE FROM {table} WHERE ctid IN (SELECT ctid FROM {table} LIMIT 1) RETURNING *;
     ```
   - This statement deletes the first record found in each specified table. The `ctid` is a system column in PostgreSQL that uniquely identifies rows within a table. By using `LIMIT 1`, only the first record (as per internal ordering) is deleted.
   - The `RETURNING *;` clause is crucial as it returns the deleted record(s), allowing the application to display what was removed.

2. **Handling of Cascade Deletes**:
   - Since foreign key constraints with `CASCADE` are set up in the database, deleting a record in one table (like `Foods` or `Dishes`) automatically triggers the deletion of related records in dependent tables (like `foodsinDish`, `Meals`, `Days`).

3. **Displaying Table Data Post-Deletion**:
   - After performing the deletions, the route executes a `SELECT * FROM {table};` query for each table to fetch the current state of the database:
     ```sql
     SELECT * FROM {table};
     ```
   - This step is vital to show the remaining data in each table, highlighting the effects of the cascade deletions and providing a clear picture of how the database's content has changed.

The SQL statements within the `/db_deletes` route are wrapped in try-except blocks to handle any exceptions and ensure that the database connection is properly managed. The route's execution demonstrates the interconnectedness of data within a relational database and the importance of understanding cascade effects in deletion operations. It effectively illustrates the practical implications of foreign key relationships and the utility of cascade deletions in maintaining data integrity.

````python
@app.route('/db_deletes')
def db_deletes():
    conn = psycopg2.connect("postgres://food_db_msqq_user:96WkFN4LYyA6g0p8n9ykbw7GT0KQudsM@dpg-clok7g1oh6hc73bia110-a/food_db_msqq")
    cur = conn.cursor()

    def format_records_as_table(records, table_name, column_names):
        response_string = f"<h2>{table_name}</h2>"
        response_string += "<table border='1'>"
        
        # Add column headers
        response_string += "<tr>"
        for col_name in column_names:
            response_string += f"<th>{col_name}</th>"
        response_string += "</tr>"
        
        # Add table data
        for record in records:
            response_string += "<tr>"
            for info in record:
                response_string += f"<td>{info}</td>"
            response_string += "</tr>"
        response_string += "</table><br>"
        return response_string

    def query_and_format_table(cur, table_name):
        cur.execute(f'SELECT * FROM {table_name};')
        records = cur.fetchall()
        column_names = [desc[0] for desc in cur.description]
        return format_records_as_table(records, table_name, column_names)

    response_string = ""

    # Correct order to respect foreign key constraints
    deletion_order = ['Days', 'Meals', 'foodsinDish', 'Dishes', 'Foods', 'Users']

    for table in deletion_order:
        try:
            # Deleting the first record found in each table
            cur.execute(f'DELETE FROM {table} WHERE ctid IN (SELECT ctid FROM {table} LIMIT 1) RETURNING *;')
            deleted_records = cur.fetchall()
            if deleted_records:
                deleted_records_formatted = ', '.join([str(record) for record in deleted_records])
                response_string += f"<p>Deleted from {table}: {deleted_records_formatted}</p>"
            else:
                response_string += f"<p>No record deleted from {table} (not found or already deleted).</p>"
        except psycopg2.Error as e:
            response_string += f"<p>Error deleting from {table}: {e}</p>"
            conn.rollback()

    # Query and display data from each table after deletion
    for table in ["Users", "Foods", "Dishes", "foodsinDish", "Meals", "Days"]:
        response_string += query_and_format_table(cur, table)

    conn.commit()
    cur.close()
    conn.close()
    return response_string
````

## Dropping the Database

The `/db_drop` route in the project is designed to completely reset the database by dropping all tables, triggers, and other database objects. This route is crucial for scenarios where a clean slate is required, such as during redevelopment or testing phases. Here's a summary of the SQL statements used (code in snippet below):

1. **Dropping Triggers**:
   - The route starts by listing all triggers to be dropped. These triggers include those for preventing updates to names in `Foods` and `Dishes`, logging updates in `Foods`, and various deletion triggers.
   - The SQL statement for dropping each trigger is:
     ```sql
     DROP TRIGGER IF EXISTS {trigger} ON public.Foods;
     ```
   - This command ensures that the specified trigger is dropped if it exists. The `public.Foods` specification indicates the table on which the trigger is defined. Note that the table name may need to be adjusted based on the trigger's actual associated table.

2. **Dropping Tables**:
   - The route then defines a list of tables to be dropped, ordered in reverse of their creation sequence. This ordering is necessary to respect foreign key constraints.
   - For each table in the list, the following SQL command is executed:
     ```sql
     DROP TABLE IF EXISTS {table};
     ```
   - This statement drops the specified table if it exists, preventing errors if a table is already absent. It's a safeguard to ensure smooth execution of the route.

3. **Dropping Enum Type**:
   - Lastly, the route drops the enum type `meal_type` used in the database with the following SQL command:
     ```sql
     DROP TYPE IF EXISTS meal_type;
     ```
   - This command removes the custom enum type, which is necessary if it needs to be redefined or is no longer required.

After executing these SQL statements, the route commits the changes to the database and closes the connection. The `/db_drop` route effectively clears the database, removing all previously defined structures and data, thereby returning the database to its initial, empty state. This functionality is particularly useful for development and testing, allowing a fresh start or a clean environment for new implementations.

````python
@app.route('/db_drop')
def dropping():
    conn = psycopg2.connect("postgres://food_db_msqq_user:96WkFN4LYyA6g0p8n9ykbw7GT0KQudsM@dpg-clok7g1oh6hc73bia110-a/food_db_msqq")
    cur = conn.cursor()
    
    # List of triggers to drop
    triggers_to_drop = [
        'prevent_name_update_foods_trigger', 'prevent_name_update_dishes_trigger', 
        'foods_update_log_trigger', 'delete_dish_trigger', 'delete_food_trigger', 
        'delete_meal_trigger', 'delete_day_user_trigger', 'delete_day_meal_trigger'
    ]

    # Drop triggers
    for trigger in triggers_to_drop:
        cur.execute(f'DROP TRIGGER IF EXISTS {trigger} ON public.Foods;')

    # Drop tables in reverse order of creation due to foreign key constraints
    tables_to_drop = ['FoodUpdatesLog', 'Days', 'Users', 'Meals', 'foodsinDish', 'Dishes', 'Foods']

    for table in tables_to_drop:
        cur.execute('DROP TABLE IF EXISTS {};'.format(table))

    # Drop the enum type meal_type as well
    cur.execute('DROP TYPE IF EXISTS meal_type;')

    conn.commit()
    conn.close()
    return "All Tables and Triggers Successfully Dropped"
````
